# 하나은행 홈페이지 크롤링
* 날짜를 적으면 해당 날짜의 매매기준율 환율 return하는 함수

```python
import requests
from bs4 import BeautifulSoup
import pickle
import tqdm
```

```python
hana = "https://www.kebhana.com/cms/rate/wpfxd651_01i_01.do"
hana_payload = {"ajax"  : "true",
"curCd"  : "",
"tmpInqStrDt"  : "2023-01-12",
"pbldDvCd"  : "0",
"pbldSqn"  : "",
"hid_key_data"  : "",
"inqStrDt"  : "20230112",
"inqKindCd"  : "1",
"hid_enc_data"  : "",
"requestTarget"  : "searchContentDiv",}
```

```python
r = requests.post(hana, data=hana_payload)
```

```python
BeautifulSoup(r.text, 'lxml').findAll("tr")[3].findAll("td")[9].text
```

```python
def exchange(date_):
    """
        date_형식은 yyyy-mm-dd 형식으로 해주세요
    """
    hana_payload["tmpInqStrDt"] = date_
    hana_payload["inqStrDt"] = date_.replace("-", "")
    r = requests.post(hana, data=hana_payload)
    return (BeautifulSoup(r.text, 'lxml').findAll("tr")[3].findAll("td")[9].text).replace(",","")
```

```python
float(exchange("2023-01-30"))
```

```python
from datetime import date, datetime, timedelta
```

```python
date_dict = dict()
for x in tqdm.tqdm(range(365)):    
    date_dict[(date(2022, 1, 1) + timedelta(days=x)).strftime("%Y-%m-%d")] = exchange((date(2022, 1, 1) + timedelta(days=x)).strftime("%Y-%m-%d"))
print(date_dict)
```

```python
with open("./kbo_result.pkl", "rb") as f:
    kbo_result = pickle.load(f)
```

```python
for player in kbo_result:
    if player["연봉"] == "":
        player["연봉(만원)"] = 0
    elif "만원" in player["연봉"]:
        player["연봉(만원)"] = int(player["연봉"].split("만원")[0]) * 10000
    elif "달러" in player["연봉"]:
        player["연봉(만원)"] = int(player["연봉"].split("달러")[0]) * float(exchange("2023-01-01"))
```

```python
sorted([(x["선수명"], x["연봉(만원)"]) for x in kbo_result], key = lambda x: x[1], reverse = True)
```

```python
print("{:,}원".format(sum([x['연봉(만원)']for x in kbo_result]) / len(kbo_result)))
```

```python
for x in kbo_result:
    x["나이"] = 2023 - int(x["생년월일"].split("년")[0])
```

# Class

```python
class stack:
    def __init__(self):
        self.items = []
        
    def push(self, value):
        self.items.append(value)
        
    def pop(self):
        value = self.items.pop()
        if value is not None:
            return value
        else:
            print("stack is empty")
    
    def __repr__(self):
        return repr(self.items)
```

```python
stack1 = stack()
stack2 = stack()
```

```python
stack1.push(3)
```

```python
print(stack1)
```

```python
print(stack2)
```

```python
import pandas as pd
class crawling:
    def __init__(self):
        self.header = {
            "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36",
        }
    
    def get_naver_stock(self, code):
        url = f"https://finance.naver.com/item/sise_day.naver?code={code}&page=1"
        r = requests.get(url, headers=self.header)
        return pd.read_html(r.text)[0].dropna()
```

```python
crawler = crawling()
```

```python
crawler.get_naver_stock('005930')
```
