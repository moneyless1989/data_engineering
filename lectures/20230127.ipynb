# 선수 상세페이지의 정보를 Dictionary 형태로 저장

```python
from bs4 import BeautifulSoup as BS
import requests
import tqdm
```

```python
player_details = "https://www.koreabaseball.com/Record/Player/PitcherDetail/Basic.aspx?playerId={}"
```

```python
players = []
for player_id in tqdm.tqdm(playid):
#웹페이지 열기
    r = requests.get(player_details.format(player_id))
    bs = BS(r.text)
#선수 정보 테이블 접근 및 딕셔너리 저장
    players.append({x.text.split(":")[0].strip():x.text.split(":")[1].strip() for x in bs.find("div", {"class", "player_basic"}).findAll("li")})
print(players)    
```

```python
with open("./kbo_playes.pkl", "wb") as f:
    pickle.dump(players, f)
```

```python
with open("./kbo_playes.pkl", "rb") as f:
    players = pickle.load(f)
```

```python
for player in players:
    if player["연봉"] == "":
        player["연봉"] = 0
    elif "만원" in player["연봉"]:
        player["연봉"] = int(player["연봉"].split("만원")[0]) * 10000
    elif "달러" in player["연봉"]:
        player["연봉"] = int(player["연봉"].split("달러")[0]) * 1235
#   print(type(player["연봉"]))
```
