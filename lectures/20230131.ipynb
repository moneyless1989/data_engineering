# generator

```python
def fib_gene(n):
    (a, b) = (0, 1)
    for _ in range(n):
        yield a
        (a, b) = (b, a+b)
```

```python
def fibo(n):
    (a, b) = (0, 1)
    for _ in range(n):
        (a, b) = (b, a+b)
    return a
```

```python
list(fib_gene(100))[-1]
```

```python
fibo(99)
```

# decorator

```python
def func_a(func):
    print("func_a")
    func()
    def inner():
        print("inner function")
        func()
    return inner
```

```python
@func_a
def func_b():
    print("func_b start")
```

```python
func_b()
```

```python
func_a(func_b())
```
