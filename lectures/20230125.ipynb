```python
head = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"
}
```

```python
mel = requests.get("https://www.melon.com/chart/index.htm", headers = head)
```

```python
def getLyric(lid):
    mel = requests.get(f"https://www.melon.com/song/detail.htm?songId={lid}", headers = head)
    bs = BeautifulSoup(mel.text)
    lyric = bs.find("div", id="d_video_summary")
    song_name = bs.find("div", class_="song_name")
    return {"song_name":BeautifulSoup(str(song_name).replace('<strong class="none">곡명</strong>', '')).text.strip(), "lyric":BeautifulSoup(str(lyric).replace("<br/>", "\n")).text.strip()}
```

```python
# mel에서 곡 ID를 찾아 ids에 저장
findID = BeautifulSoup(mel.text)
tags = findID.findAll("tr", id="lst50")
tags.extend(findID.findAll("tr", id="lst100"))
ids = [x["data-song-no"] for x in tags]
# 각 ID로 get을 돌리고 가사를 찾음
# 각 가사를 리스트나 딕셔너리에 저장
lyrics = []
for i in ids:
    lyrics.append({"id":i, "song_name":getLyric(i)["song_name"], "lyric":getLyric(i)["lyric"]})
```

```python
lyrics[0]
```

```python
os.mkdir("./song")
```

```python
#가사를 파일에 쓰기
def song_write(song):
    song_file = open(f"./song/{song['song_name']}.txt", "w", encoding="UTF-8")
    song_file.write(song["lyric"])
    song_file.close()
    return 0
```

```python
for i in lyrics:
    song_write(i)
```

```python
r = requests.get("https://www.ppomppu.co.kr/zboard/zboard.php?id=ppomppu")
bs = BeautifulSoup(r.text)
bs2 = bs.find("table", id='revolution_main_table').findAll("tr")
for x in bs2:
    try:
        print(x.findAll("a")[-1].text)
    except:
        print("해당번호에는 데이터 없음")
```

```python
total = []
for idx, x in enumerate(bs2):
    try:
        #print(idx, x.findAll("a")[-1].text)
        total.append(x.findAll("a")[-1].text)
    except:
        #print(idx, "해당번호에는 데이터 없음")
        total.append("해당번호에는 데이터 없음")
print(list(set(total)))
```

```python
import re
p = re.compile("([0-9]+[,][0-9]+)")


total2 = []
for x in list(set(total)):
    try:
        #print(f"{x} -- {p.findall(x)[0]}" )
        total2.append(p.findall(x)[0])
    except:
        pass
```
